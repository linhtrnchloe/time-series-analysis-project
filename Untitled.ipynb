PART 3: INTERPRETING THE DIAGNOSTICS OF GARCH AND EGARCH

```python
import statsmodels.api as sm

def sign_bias_test(std_resid):
    z  = std_resid.dropna()
    z2 = (z**2).values[1:]
    zl = z.shift(1).dropna().values
    
    neg      = (zl < 0).astype(float)
    pos      = (zl > 0).astype(float)
    neg_size = neg * zl
    pos_size = pos * zl
    
    X   = np.column_stack([np.ones_like(zl), neg, neg_size, pos_size])
    ols = sm.OLS(z2, X).fit()
    
    print("Sign Bias Test Results:")
    print(f"{'Test':<20} {'t-stat':>10} {'p-value':>10}")
    print("-" * 42)
    print(f"{'Sign Bias':<20} {ols.tvalues[1]:>10.4f} {ols.pvalues[1]:>10.4f}")
    print(f"{'Negative Size Bias':<20} {ols.tvalues[2]:>10.4f} {ols.pvalues[2]:>10.4f}")
    print(f"{'Positive Size Bias':<20} {ols.tvalues[3]:>10.4f} {ols.pvalues[3]:>10.4f}")

sign_bias_test(garch11_fitted.std_resid)
```

```python
print("=" * 50)
print("SIGN BIAS TEST - GARCH(1,1)")
print("=" * 50)
sign_bias_test(garch11_fitted.std_resid)

print()
print("=" * 50)
print("SIGN BIAS TEST - EGARCH(1,1)")
print("=" * 50)
sign_bias_test(egarch_fitted.std_resid)
```

```python
from scipy import stats

def pearson_gof(std_resid, nu, bins=20):
    z   = std_resid.dropna().values
    u   = stats.t.cdf(z/np.sqrt((nu-2)/nu), df=nu)
    obs, _ = np.histogram(u, bins=np.linspace(0, 1, bins+1))
    exp    = len(u)/bins
    stat   = ((obs-exp)**2/exp).sum()
    dof    = bins-1
    p      = 1-stats.chi2.cdf(stat, dof)
    print(f"Pearson GoF Statistic : {stat:.4f}")
    print(f"Degrees of freedom    : {dof}")
    print(f"P-value               : {p:.4f}")

print("=" * 50)
print("PEARSON GOF TEST - GARCH(1,1)")
print("=" * 50)
nu_garch = garch11_fitted.params["nu"]
pearson_gof(garch11_fitted.std_resid, nu_garch)

print()
print("=" * 50)
print("PEARSON GOF TEST - EGARCH(1,1)")
print("=" * 50)
nu_egarch = egarch_fitted.params["nu"]
pearson_gof(egarch_fitted.std_resid, nu_egarch)
```

```python
from scipy.special import gammaln

def nyblom_test(fitted, data):
    params = fitted.params.values
    k = len(params)
    x = np.asarray(data, float)
    T = len(x)
    
    mu    = params[0]
    omega = params[1]
    alpha = params[2]
    beta  = params[3]
    nu    = params[4]
    
    # compute per observation log likelihood contributions
    e  = x - mu
    s2 = np.empty(T)
    s2[0] = np.var(e)
    for t in range(1, T):
        s2[t] = omega + alpha*e[t-1]**2 + beta*s2[t-1]
    
    z = e / np.sqrt(s2)
    c = gammaln((nu+1)/2) - gammaln(nu/2) - 0.5*np.log(np.pi*(nu-2))
    ll_t = c - ((nu+1)/2)*np.log(1 + z**2/(nu-2)) - 0.5*np.log(s2)
    
    # numerical scores via finite differences
    S = np.zeros((T, k))
    for j in range(k):
        step = 1e-4 * max(1, abs(params[j]))
        pp = params.copy(); pm = params.copy()
        pp[j] += step; pm[j] -= step
        
        def ll_contribs(p):
            mu_,om_,al_,be_,nu_ = p
            e_  = x - mu_
            s2_ = np.empty(T); s2_[0] = np.var(e_)
            for t in range(1, T):
                s2_[t] = om_ + al_*e_[t-1]**2 + be_*s2_[t-1]
            z_  = e_ / np.sqrt(s2_)
            c_  = gammaln((nu_+1)/2)-gammaln(nu_/2)-0.5*np.log(np.pi*(nu_-2))
            return c_ - ((nu_+1)/2)*np.log(1+z_**2/(nu_-2)) - 0.5*np.log(s2_)
        
        S[:, j] = (ll_contribs(pp) - ll_contribs(pm)) / (2*step)
    
    # Nyblom statistic
    Sc  = S - S.mean(0)
    Omega = (Sc.T @ Sc) / T
    cum = np.cumsum(Sc, 0)
    M   = (cum.T @ cum) / T**2
    joint = np.trace(np.linalg.solve(Omega, M))
    indiv = np.diag(M) / np.diag(Omega)
    
    print("Nyblom Stability Test")
    print("=" * 45)
    print(f"{'Parameter':<15} {'Statistic':>10} {'5% Crit':>10}")
    print("-" * 45)
    for i, name in enumerate(fitted.params.index):
        flag = "❌ unstable" if indiv[i] > 0.47 else "✅ stable"
        print(f"{name:<15} {indiv[i]:>10.4f} {0.47:>10.2f}  {flag}")
    print("-" * 45)
    print(f"{'Joint':<15} {joint:>10.4f} {1.24:>10.2f}  {'❌ unstable' if joint > 1.24 else '✅ stable'}")

print("NYBLOM TEST - GARCH(1,1)")
nyblom_test(garch11_fitted, in_sample.values)
```

PART 4: CALCULATING VAR

```python
from scipy import stats

def calculate_var(conditional_vol, mu, nu, alpha):
    # Student-t quantile (standardized)
    q = stats.t.ppf(alpha, nu) * np.sqrt((nu-2)/nu)
    var = mu + conditional_vol * q
    return var

# Parameters
mu_garch  = garch11_fitted.params["mu"]
nu_garch  = garch11_fitted.params["nu"]
mu_egarch = egarch_fitted.params["mu"]
nu_egarch = egarch_fitted.params["nu"]

# In-sample VaR for both confidence levels
var_garch_95  = calculate_var(garch11_fitted.conditional_volatility, 
                               mu_garch, nu_garch, 0.05)
var_garch_99  = calculate_var(garch11_fitted.conditional_volatility, 
                               mu_garch, nu_garch, 0.01)
var_egarch_95 = calculate_var(egarch_fitted.conditional_volatility, 
                               mu_egarch, nu_egarch, 0.05)
var_egarch_99 = calculate_var(egarch_fitted.conditional_volatility, 
                               mu_egarch, nu_egarch, 0.01)

# Plot
fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# GARCH
axes[0].plot(in_sample.index, in_sample.values, 
             color="grey", lw=0.7, label="Returns")
axes[0].plot(var_garch_95.index, var_garch_95.values, 
             color="blue", lw=1, label="VaR 95%")
axes[0].plot(var_garch_99.index, var_garch_99.values, 
             color="red", lw=1, label="VaR 99%")
axes[0].set_title("GARCH(1,1) - In-Sample VaR")
axes[0].legend()

# EGARCH
axes[1].plot(in_sample.index, in_sample.values, 
             color="grey", lw=0.7, label="Returns")
axes[1].plot(var_egarch_95.index, var_egarch_95.values, 
             color="blue", lw=1, label="VaR 95%")
axes[1].plot(var_egarch_99.index, var_egarch_99.values, 
             color="red", lw=1, label="VaR 99%")
axes[1].set_title("EGARCH(1,1) - In-Sample VaR")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
out_of_sample = port_pct[port_pct.index >= "2024-05-01"]

print(f"Out-of-sample observations: {len(out_of_sample)}")
print(f"From: {out_of_sample.index.min().date()} to {out_of_sample.index.max().date()}")
```

```python
history = list(in_sample.values)

var_garch_oos_95  = []
var_garch_oos_99  = []
var_egarch_oos_95 = []
var_egarch_oos_99 = []

for i, date in enumerate(out_of_sample.index):
    
    # fit both models on expanding window
    s = pd.Series(history)
    
    res_g = arch_model(s, mean="Constant", vol="GARCH", 
                       p=1, q=1, dist="t").fit(disp="off")
    res_e = arch_model(s, mean="Constant", vol="EGARCH", 
                       p=1, o=1, q=1, dist="t").fit(disp="off")
    
    # one step ahead forecast
    f_g = res_g.forecast(horizon=1, reindex=False)
    f_e = res_e.forecast(horizon=1, reindex=False)
    
    sig_g = np.sqrt(f_g.variance.values[-1, 0])
    sig_e = np.sqrt(f_e.variance.values[-1, 0])
    
    mu_g  = res_g.params["mu"]
    nu_g  = res_g.params["nu"]
    mu_e  = res_e.params["mu"]
    nu_e  = res_e.params["nu"]
    
    var_garch_oos_95.append(calculate_var(sig_g, mu_g, nu_g, 0.05))
    var_garch_oos_99.append(calculate_var(sig_g, mu_g, nu_g, 0.01))
    var_egarch_oos_95.append(calculate_var(sig_e, mu_e, nu_e, 0.05))
    var_egarch_oos_99.append(calculate_var(sig_e, mu_e, nu_e, 0.01))
    
    # add actual return to history
    history.append(out_of_sample.iloc[i])
    
    if i % 50 == 0:
        print(f"Progress: {i}/{len(out_of_sample)} days done")

print("Done!")
```

```python
from scipy import stats

def kupiec_test(breaches, alpha, n):
    x  = int(breaches.sum())
    pi = x/n
    if x == 0:
        return np.nan, 1.0
    LR = -2*((n-x)*np.log(1-alpha) + x*np.log(alpha)
             -(n-x)*np.log(1-pi)   - x*np.log(pi))
    p  = 1 - stats.chi2.cdf(LR, 1)
    return LR, p

def christoffersen_test(breaches):
    v    = breaches.astype(int).values
    n00=n01=n10=n11=0
    for i in range(1, len(v)):
        a,b = v[i-1], v[i]
        if   a==0 and b==0: n00+=1
        elif a==0 and b==1: n01+=1
        elif a==1 and b==0: n10+=1
        else:                n11+=1
    pi01 = n01/(n00+n01) if (n00+n01)>0 else 0
    pi11 = n11/(n10+n11) if (n10+n11)>0 else 0
    pi   = (n01+n11)/(n00+n01+n10+n11)
    def sf(x): return x if x>0 else 1e-300
    L0 = sf(1-pi)**(n00+n10) * sf(pi)**(n01+n11)
    L1 = sf(1-pi01)**n00 * sf(pi01)**n01 * sf(1-pi11)**n10 * sf(pi11)**n11
    LR = -2*np.log(L0/L1)
    p  = 1-stats.chi2.cdf(LR, 1)
    return LR, p

print(f"{'Model':<12} {'Level':>6} {'Kupiec p':>10} {'Christoff p':>12} {'Verdict':>15}")
print("-" * 58)
for col in breaches.columns:
    model, level = col.split("_")
    alpha  = 0.05 if level == "95" else 0.01
    LRk,pk = kupiec_test(breaches[col], alpha, n)
    LRc,pc = christoffersen_test(breaches[col])
    verdict = "✅ Pass" if pk>0.05 and pc>0.05 else "❌ Fail"
    print(f"{model:<12} {level+'%':>6} {pk:>10.4f} {pc:>12.4f} {verdict:>15}")
```
